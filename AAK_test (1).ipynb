```python
df_skills_cleaned = df_skills.dropna()
```

* dropping in silver layer

```python
df_skills_cleaned.to_csv(silver_path / 'df_skills_silver.csv')
```

```python
df_jobs_p_cleaned.to_csv(silver_path / 'df_postings_silver.csv')
```

```python
df_joined = df_jobs_p_cleaned.merge(df_skills_cleaned, on = 'job_link' , how = 'inner' )
```

```python
df_joined
```

```python
df_joined.to_csv(gold_path / 'df_joined_gold.csv')
```

```python
df_joined['first_seen'] = pd.to_datetime(df_joined['first_seen'], errors='coerce' , utc=True)

df_joined_golden = df_joined.dropna(subset=['first_seen']).copy()
```

```python
df_joined_golden['week_start'] = (
            df_joined_golden['first_seen']
            .dt.tz_convert(None)
            .dt.to_period('W-MON')
            .apply(lambda p: p.start_time)
)

df_joined_golden['year_week'] = df_joined_golden['week_start'].dt.strftime("%G-W%V")
```

```python
weekly_counts = (
    df_joined_golden.groupby("week_start", as_index=False).size().rename(columns={"size" : "postings"}).sort_values("week_start")
)

weekly_counts["postings_4wk_ma"] = weekly_counts["postings"].rolling(4, min_periods = 1 ).mean()
```

```python
print(weekly_counts)
```

```python



levels_count = df_joined_golden['job_level'].value_counts(dropna=False)
level_pct = df_joined_golden['job_level'].value_counts(normalize=True , dropna=False).mul(100).round(2)

dist_job_levels = (
                   pd.DataFrame({'count' : levels_count, 'pct' : level_pct})
                   .reset_index(names=['job_level'])
                   .sort_values('count', ascending=False , ignore_index=True)

)

print(dist_job_levels)
```

```python



types_count = df_joined_golden['job_type'].value_counts(dropna=False)
type_pct = df_joined_golden['job_type'].value_counts(normalize=True , dropna=False).mul(100).round(2)

dist_job_types = (
                   pd.DataFrame({'count' : types_count, 'pct' : type_pct})
                   .reset_index(names=['job_type'])
                   .sort_values('count', ascending=False , ignore_index=True)

)

print(dist_job_types)
```
